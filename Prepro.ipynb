```python
import pandas as pd
import os
import time
import glob
```

```python
count = 0
data = []
start = time.time()
for file in glob.glob('Data/Stocks/*.txt'):
    if(os.stat(file).st_size != 0): # if the file is NOT empty
        tempdf = pd.read_csv(file)
        tempdf['Stock'] = str.upper(file[12:-7:1]) # Stock Name
        tempdf = tempdf[['Date','Open','High','Low','Close','Volume','Stock']] # we keep the columns that we want 
        data.append(tempdf)        
    count += 1
    if count%1000 == 0:
        print('Progress:',count,'/ 7195 = ',int(100*count/7195),'%  Time = ', int(time.time() - start)//60,'min')  
print(count)
data = pd.concat(data)
data.reset_index(inplace=True, drop=True)
```

```python
# print(data)
df = data.sort_values(by=['Date'], axis=0)
```
